```python
def Hotel_info():
    clm={'id':[],'Listing_url':[],'Name':[],'Description':[],'Neighborhood_overview':[],'Street':[],'Suburb':[],'Market':[],'Longitude':[],
            'Latitude':[],'Country':[],'Country_code':[],'Price':[],'Weekly_price':[],'Monthly_price':[],'Security_deposit':[],
            'Cancelation_policy':[],'Cleaning_fee':[]}
    for i in airbnb_data:
        ids=i['_id']
        clm['id'].append(ids)
        listingurl=i["listing_url"]
        clm['Listing_url'].append(listingurl)
        name=i['name']
        clm['Name'].append(name)
        descr=i["description"]
        clm['Description'].append(descr)
        neighbor=i['neighborhood_overview']
        clm["Neighborhood_overview"].append(neighbor)
        Street=i['address']['street']
        clm['Street'].append(Street)
        suburb=i['address']['suburb']
        clm['Suburb'].append(suburb)
        market=i['address']['market']
        clm['Market'].append(market)
        longitude=i['address']['location']['coordinates'][0]
        clm['Longitude'].append(longitude)
        latitude=i['address']['location']['coordinates'][1]
        clm['Latitude'].append(latitude)
        country=i['address']['country']
        clm['Country'].append(country)
        countrycode=i['address']['country_code']
        clm['Country_code'].append(countrycode)
        price=i['price']
        clm['Price'].append(price)
        weekprice=i.get('weekly_price')
        clm["Weekly_price"].append(weekprice)
        monthprice=i.get('monthly_price')
        clm["Monthly_price"].append(monthprice)
        security=i.get('security_deposit')
        clm['Security_deposit'].append(security)
        cancel=i['cancellation_policy']
        clm['Cancelation_policy'].append(cancel)
        cleaning=i.get('cleaning_fee')
        clm["Cleaning_fee"].append(cleaning)

    

    Hotel_data=pd.DataFrame(clm)
    Hotel_data['Security_deposit']=Hotel_data['Security_deposit'].fillna(0)
    Hotel_data['Cleaning_fee']=Hotel_data['Cleaning_fee'].fillna(0)
    Hotel_data['Weekly_price']=Hotel_data['Weekly_price'].fillna(0)
    Hotel_data['Monthly_price']=Hotel_data['Monthly_price'].fillna(0)
    Hotel_data['Price']=Hotel_data['Price'].astype(dtype='int64')
    Hotel_data['Security_deposit']=Hotel_data['Security_deposit'].astype(dtype='int64')
    Hotel_data['Cleaning_fee']=Hotel_data['Cleaning_fee'].astype(dtype='int64')
    Hotel_data['Weekly_price']=Hotel_data['Weekly_price'].astype(dtype='int64')
    Hotel_data['Monthly_price']=Hotel_data['Monthly_price'].astype(dtype='int64')


    return Hotel_data



```

```python
def Room_info():
    clm1={"id":[],"Images":[],"Property_Type":[],"Room_Type":[],"Bed_Type":[],"Minimum_Nights":[],"Maximum_Nights":[],"Accomodates":[],"Bedrooms":[],
      'Beds':[],'Bathrooms':[],'Guests_included':[],'Amenities':[],'Availability_30':[],'Availability_60':[],'Availability_90':[],
      'Availability_365':[]}
    for i in airbnb_data:
        ids=i['_id']
        clm1['id'].append(ids)
        image=i["images"]['picture_url']
        clm1["Images"].append(image)
        property_types=i['property_type']
        clm1['Property_Type'].append(property_types)
        room=i['room_type']
        clm1['Room_Type'].append(room)
        bed=i['bed_type']
        clm1['Bed_Type'].append(bed)
        minnight=i['minimum_nights']
        clm1['Minimum_Nights'].append(minnight)
        maxnight=i['maximum_nights']
        clm1['Maximum_Nights'].append(maxnight)
        accomod=i['accommodates']
        clm1['Accomodates'].append(accomod)
        bedroom=i.get('bedrooms')
        clm1['Bedrooms'].append(bedroom)
        beds=i.get('beds')
        clm1['Beds'].append(beds)
        bathroom=i.get('bathrooms')
        clm1['Bathrooms'].append(bathroom)
        guest=i.get('guests_included')
        clm1["Guests_included"].append(guest)
        ammenity=i['amenities']
        clm1['Amenities'].append(ammenity)
        avail_30=i['availability']["availability_30"]
        clm1['Availability_30'].append(avail_30)
        avail_60=i['availability']["availability_60"]
        clm1['Availability_60'].append(avail_60)
        avail_90=i['availability']["availability_90"]
        clm1['Availability_90'].append(avail_90)
        avail_365=i['availability']["availability_365"]
        clm1['Availability_365'].append(avail_365)

    Room_data=pd.DataFrame(clm1)


    Room_data["Beds"]=Room_data['Beds'].fillna(0)
    Room_data["Bedrooms"]=Room_data['Bedrooms'].fillna(0)
    Room_data["Bathrooms"]=Room_data['Bathrooms'].fillna(0)
    Room_data["Beds"]=Room_data['Beds'].astype(dtype='int64')
    Room_data["Bedrooms"]=Room_data["Bedrooms"].astype(dtype='int64')
    Room_data["Bathrooms"]=Room_data["Bathrooms"].astype(dtype='int64')
    Room_data["Guests_included"]=Room_data["Guests_included"].astype(dtype='int64')
    Room_data['Amenities']=[', '.join(str(a) for a in x) if isinstance(x, list) else ' ' for x in Room_data['Amenities']]   
    return Room_data

```

```python
def Host_info():
      clm2={'id':[],'Host_Id':[],'Host_Name':[],'Host_location':[],'Host_about':[],'Host_verification':[],'Host_Neighbourhood':[],
      'Host_total_list':[],'Host_response_rate':[]}
      for i in airbnb_data:
            id=i['_id']
            clm2["id"].append(id)
            hostid=i['host']['host_id']
            clm2['Host_Id'].append(hostid)
            hostname=i['host']['host_name']
            clm2['Host_Name'].append(hostname)
            host_loc=i['host']['host_location']
            clm2["Host_location"].append(host_loc)
            host_abt=i['host']['host_about']
            clm2["Host_about"].append(host_abt)
            hostver=i['host']['host_verifications']
            clm2['Host_verification'].append(hostver)
            neighbour=i['host']['host_neighbourhood']
            clm2['Host_Neighbourhood'].append(neighbour)
            host_total_list=i['host']['host_total_listings_count']
            clm2["Host_total_list"].append(host_total_list)
            host_res=i['host'].get('host_response_rate')
            clm2["Host_response_rate"].append(host_res)
      Host_data=pd.DataFrame(clm2)
      Host_data['Host_response_rate']=Host_data['Host_response_rate'].fillna(0).astype(dtype='int64')
      Host_data['Host_verification']=[', '.join(map(str,x)) if isinstance else ' ' for x in Host_data['Host_verification']]

      return Host_data

```

```python
def Review_info():
    clm3={"id":[],"Number_of_review":[],"First_review":[],"Last_review":[],"Overall_score":[],"Cleanliness_score":[],
      "Checkin_score":[],"Communication_score":[],"Location_score":[],"Rating":[]}
    for i in airbnb_data:
        id=i["_id"]
        clm3["id"].append(id)
        review=i['number_of_reviews']
        clm3['Number_of_review'].append(review)
        firstrev=i.get('first_revies')
        clm3['First_review'].append(firstrev)
        lastrev=i.get('last_revies')
        clm3['Last_review'].append(lastrev)
        overall=i["review_scores"].get("review_score_accuracy")
        clm3["Overall_score"].append(overall)
        clean=i["review_scores"].get("review_scores_cleanliness")
        clm3["Cleanliness_score"].append(clean)
        checkin=i["review_scores"].get('review_scores_checkin')
        clm3["Checkin_score"].append(checkin)
        comm=i["review_scores"].get("review_scores_communication")
        clm3["Communication_score"].append(comm)
        location=i["review_scores"].get("review_score_location")
        clm3['Location_score'].append(location)
        rating=i["review_scores"].get("review_score_rating")
        clm3['Rating'].append('rating')
    Review_data=pd.DataFrame(clm3)
    Review_data['First_review']=Review_data['First_review'].fillna('Mot Available')
    Review_data['Last_review']=Review_data['Last_review'].fillna('Not Available')
    Review_data["Cleanliness_score"]=Review_data["Cleanliness_score"].fillna(0)
    Review_data["Checkin_score"]=Review_data["Checkin_score"].fillna(0)
    Review_data["Communication_score"]=Review_data["Communication_score"].fillna(0)
    Review_data["Location_score"]=Review_data["Location_score"].fillna(0)
    Review_data["Cleanliness_score"]=Review_data["Cleanliness_score"].astype(dtype='int64')
    Review_data["Checkin_score"]=Review_data["Checkin_score"].astype(dtype='int64')
    Review_data["Communication_score"]=Review_data["Communication_score"].astype(dtype='int64')
    Review_data["Location_score"]=Review_data["Location_score"].astype(dtype='int64')
    return Review_data

```

```python
def Reviewer_info():
        clm4={"id":[],"Comment_id":[],"Date":[],"Reviewer_id":[],"Reviewer_name":[],"Comments":[]}
        for j in airbnb_data:
                r_list=j.get('review',[])
                for i in r_list:
                        id=i.get('listing_id')
                        clm4["id"].append(id)
                        comment_id=i.get('_id')
                        clm4["Comment_id"].append(comment_id)
                        date=i.get('date')
                        clm4["Date"].append(date)
                        reviewer_id=i.get('reviewer_id')
                        clm4["Reviewer_id"].append(reviewer_id)
                        reviewer_name=i.get('reviewer_name')
                        clm4["Reviewer_name"].append(reviewer_name)
                        comments=i.get('comments')
                        clm4["Comments"].append(comments)
        Reviewer_data=pd.DataFrame(clm4)
        Reviewer_data["Reviewer_name"]=Reviewer_data["Reviewer_name"].fillna('Not Available')
        Reviewer_data["Comments"]=Reviewer_data["Comments"].fillna('Not Available')
        return Reviewer_data
        
```

```python
mycursor.execute('''create table if not exists hotels_info (id VARCHAR(50) PRIMARY KEY, Listing_url VARCHAR(500),
                Name VARCHAR(100), Description TEXT, Neighborhood_overview TEXT, Street VARCHAR(100),
                Suburb VARCHAR(100), Market VARCHAR(100), Longitude FLOAT(10), Latitude FLOAT(10),Country VARCHAR(100),
                Country_code VARCHAR(50), Price INT,  Weekly_price INT, Monthly_price INT, Security_deposit INT,
                Cancelation_policy VARCHAR(50),Cleaning_fee INT)
                    ''')
for index,row in Hotel_info().iterrows():
   mycursor.execute('''INSERT INTO hotels_info (id, Listing_url, Name, Description, Neighborhood_overview, Street,Suburb,
                  Market, Longitude,Latitude, Country, Country_code, Price, Weekly_price, Monthly_price,Security_deposit,
                  cancelation_policy,cleaning_fee) 
                  VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)''',
                  (row['id'],row['Listing_url'],row['Name'],row['Description'],row['Neighborhood_overview'],row['Street'],
                   row['Suburb'],row['Market'],row['Longitude'],row['Latitude'],row['Country'],row['Country_code'],
                   row['Price'],row['Weekly_price'],row['Monthly_price'],row['Security_deposit'],row['Cancelation_policy'],
                   row['Cleaning_fee']))


mydb.commit()
```

```python
mycursor.execute('''create table if not exists rooms_info (id VARCHAR(50), Images VARCHAR(2000), Property_Type VARCHAR(50),
                Room_Type VARCHAR(50), Bed_Type VARCHAR(50), Minimum_Nights VARCHAR(10), Maximum_Nights VARCHAR(10),
                Accomodates INT, Bedrooms INT, Beds INT, Bathrooms INT,Guests_included INT,Amenities TEXT,
                Availability_30 INT,Availability_60 INT,Availability_90 INT,Availability_365 INT,
                FOREIGN KEY(id) REFERENCES hotels_info(id))''')
for index,row in Room_info().iterrows():
    mycursor.execute('''INSERT INTO rooms_info (id,Images,Property_Type,Room_Type,Bed_Type,Minimum_Nights,Maximum_Nights,
                     Accomodates,Bedrooms,Beds,Bathrooms,Guests_included,Amenities,Availability_30,Availability_60,
                     Availability_90,Availability_365)
                     VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)''',
                    (row['id'],row['Images'],row['Property_Type'],row['Room_Type'],row['Bed_Type'],row['Minimum_Nights'],row['Maximum_Nights'],
                     row['Accomodates'],row['Bedrooms'],row['Beds'],row['Bathrooms'],row['Guests_included'],row['Amenities'],
                     row['Availability_30'],row['Availability_60'],row['Availability_90'],row['Availability_365']))

mydb.commit()
```
